# Introduction to File Handling in Python

**File handling** is an important part of any programming language, as it allows you to store, retrieve, and manipulate data in files permanently (instead of losing it when the program ends).

In Python, file handling is done primarily using the built-in `open()` function.

## Opening a File 
```python 
file_object = open("filename", "mode")
```
- `filename` ‚Üí name or path of the file (e.g., "data.txt" or "C:/folder/data.txt")
- `mode` ‚Üí specifies what you want to do with the file (read/write/append, etc.)

### Common Modes 
| Mode   | Meaning            | Description                                                      |
| ------ | ------------------ | ---------------------------------------------------------------- |
| `"r"`  | Read               | Opens file for reading (default). Error if file doesn‚Äôt exist.   |
| `"w"`  | Write              | Opens file for writing. Creates new file or overwrites existing. |
| `"a"`  | Append             | Opens file for appending. Adds data to the end.                  |
| `"x"`  | Exclusive Creation | Creates a new file. Error if file already exists.                |
| `"t"`  | Text Mode          | Default. Reads/writes text data.                                 |
| `"b"`  | Binary Mode        | Reads/writes binary data (images, videos).                       |
| `"r+"` | Read + Write       | Opens file for both reading and writing.                         |


````python
f = open('./Files/sample.txt', 'r')
print(f.read())
f.close()
````

## Reading a File 

Python gives multiple methods to read file contents:
### (a) `read()` ‚Üí Reads entire file

````python
f = open("./Files/sample.txt", "r")
print(f.read())   # prints everything
f.close()
````

### (b) `read(n)` ‚Üí Reads n characters

````python
f = open("./Files/sample.txt", "r")
print(f.read(61))   # prints first 61 characters
f.close()
````

### (c) `readline()` ‚Üí Reads one line at a time

````python
f = open("./Files/sample.txt", "r")
print(f.readline())   # prints first line
print(f.readline())   # prints second line
f.close()
````

### (d) `readlines()` ‚Üí Reads all lines into a list

````python
f = open("./Files/sample.txt", "r")
for line in f.readlines():
    print(line)
f.close()
````

Note, that we are getting an extra line between the lines. The `\n` extra is coming from the `.txt` file at the end of the first line.

````python
f = open("./Files/sample.txt", "r")
for line in f.readlines():
    print(line.strip())
f.close()
````

## Writing a file 

### (a) `write()` ‚Üí Write string to file.
- Creates a new file if it does not exist!

````python
f = open("./Files/sample.txt", "w")   # overwrites file
f.write("Python is dynamically type-checked and garbage-collected.\n")
f.write("It supports multiple programming paradigms, including structured (particularly procedural), object-oriented and functional programming.")
f.close()
````

````python
# Let us read the lines now 
f = open("./Files/sample.txt", "r")
for line in f.readlines():
    print(line.strip())
f.close()
````

### (b) `writelines()` ‚Üí Write multiple lines
- Pass the lines inside a list!

````python
f = open("./Files/sample.txt", "w")   # overwrites file
f.writelines(
    [
        "Python is dynamically type-checked and garbage-collected.\n",
        "It supports multiple programming paradigms, including structured (particularly procedural), object-oriented and functional programming.\n"
    ]
)
f.close()
````

````python
# Let us read the lines now
f = open("./Files/sample.txt", "r")
for line in f.readlines():
    print(line.strip())
f.close()
````

## Closing a File

- Always close files after use to free resources.
- But instead of manually closing, best practice is using with statement:

````python
with open('./Files/sample.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        print(line.strip())
# file auto closed here
````

## Working with Binary Files

For images, audio, or videos:

## Reading an image file 

````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import io

with open('./Files/image.png', 'rb') as f:
    content = f.read()

# convert bytes to image 
img = mpimg.imread(io.BytesIO(content), format='png')

plt.imshow(img)
plt.axis('off')
plt.show()
````

üëç Let‚Äôs break down this line:

```python
img = mpimg.imread(io.BytesIO(content), format='png')
```

#### üîπ Step by Step Explanation

1. **`content`**

   * This is the raw **binary data** of your image.
   * You got it from:

     ```python
     with open("image.png", "rb") as f:
         content = f.read()
     ```
   * At this point, `content` is just bytes (`b'....'`).

2. **`io.BytesIO(content)`**

   * Wraps the raw bytes into a **file-like object** that behaves like a real file.
   * `imread` expects something that looks like a file (not just bytes).
   * So `BytesIO` acts like a *virtual file in memory*.

   ‚úÖ Example:

   ```python
   import io
   fake_file = io.BytesIO(content)
   ```

3. **`mpimg.imread(...)`**

   * `mpimg` is `matplotlib.image`.
   * `imread` reads an image from a file (or file-like object) and returns it as a **NumPy array**.
   * This array contains the pixel values of your image.

     * For grayscale ‚Üí 2D array
     * For RGB ‚Üí 3D array (height √ó width √ó 3)
     * For RGBA ‚Üí 3D array (height √ó width √ó 4)

4. **`format='png'`**

   * Explicitly tells `imread` the image format (in this case PNG).
   * It‚Äôs useful when reading from a `BytesIO` object because it doesn‚Äôt have a filename (so `imread` can‚Äôt guess the format from the extension like `.png` or `.jpg`).
   * Without it, you may get errors for certain formats.


#### ‚úÖ So in plain words:

This line takes the **binary image data (`content`)**, turns it into a **file-like object (`BytesIO`)**, and then `matplotlib` reads it as if it was a real `.png` file ‚Üí returning a **NumPy array of pixels**.

## File Pointer and seek()

The file object maintains a cursor (pointer).
- `tell()` ‚Üí shows current cursor position
- `seek(pos)` ‚Üí move cursor to position pos

````python
f = open('./Files/sample.txt', 'r')
print(f.read(57))
print(f'The cursor position is at : {f.tell()}')
f.seek(0)              # reset cursor
print(f.read())        # read full again
f.close()
````

## Hybrid File Modes in Python

| Mode   | Meaning       | Behavior                                                                                                                      |
| ------ | ------------- | ----------------------------------------------------------------------------------------------------------------------------- |
| `"r+"` | Read + Write  | Opens file for both reading and writing. File must exist. Cursor starts at beginning.                                         |
| `"w+"` | Write + Read  | Creates a new file (or overwrites existing). Allows reading too. Cursor starts at beginning.                                  |
| `"a+"` | Append + Read | Opens file for appending and reading. File is created if it doesn‚Äôt exist. Cursor at end for writing, but you can still read. |



````python
with open("./Files/sample.txt", "r+") as f:
    content = f.read()
    print("Before:", content)
    f.write("\nNew line added!")   # writes at end (since cursor moved after read)
    f.seek(0)
    print("After:", f.read())
````

‚ö†Ô∏è File must already exist.

````python
with open("./Files/sample.txt", "w+") as f:
    f.writelines(
        [
            "Python is a high-level, general-purpose programming language.\n"
        ]
    )
    f.seek(0)      # reset cursor to start
    print(f.read())  # can read what was written
````

‚ö†Ô∏è Destroys old content if file exists.

````python
with open("./Files/sample.txt", "a+") as f:
    f.write("Its design philosophy emphasizes code readability with the use of significant indentation.\n")
    f.seek(0)    # move cursor to start
    print(f.read())   # read whole file including appended content
````
